```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("coffee quality data.csv") #loading the data
df
df.info() #basic info and summary statistics
df.isnull().sum() #check for missing values
df.describe() #summary statistics
df.dropna(inplace=True)#drop rows with missing values
%matplotlib inline
plt.xlabel('Altitude')
plt.ylabel('Total Cup Points')
plt.scatter(df['Altitude'],df['Total Cup Points'],color='red',marker='+')
df.hist(bins=15,figsize=(15,8))
plt.suptitle("Distribution of features") #supertitle for entire figure
plt.show()
sns.pairplot(df) # Pairplot to visualize relationships between features
df
from sklearn.linear_model import LinearRegression
# Define the regression model
reg = LinearRegression()
# Fit the model
X = df[['Altitude']]  # This should already be a 2D array
y = df['Total Cup Points']  # This is a 1D array
reg.fit(X, y)
reg.predict([1200])
```
